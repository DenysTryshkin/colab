В шаблоні наведені числові вектори $\overline{a}$ та $\overline{b}$. Порахуй наступне:
- сума $\overline{a}$ та $\overline{b}$
- різниця $\overline{a}$ та $\overline{b}$
- сума $\overline{a}$ та $\overline{b}^T$. Поясни отриманий результат.
- матричний добуток (dot product) $\overline{a}$ та $\overline{b}^T$.
- матричний добуток (dot product) $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- добуток Адамара (Hadamard product) $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- ділення $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- ділення $\overline{a}$ та $\overline{b}^T$. Поясни отриманий результат.

````python
import numpy as np
# Код оголошення векторів не можна змінювати
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])
````

````python
import numpy as np
# Код оголошення векторів не можна змінювати
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])
res = a + b
print(res)
````

````python
res = a - b
print(res)
````

````python
# Пояснення: Транспонування матриці - це операція над матрицею, при якій її рядки та стовпці міняються місцями.
# Кожен елемент вектора a додається до кожного елемента відповідного рядка вектора b_T, що призводить до утворення нової матриці розміром 5×5.
res = a + b.T
print(res)
````

````python
res = a.dot(b.T)
print(res)
````

````python
# Пояснення: Щоб обчислити матричний добуток, потрібно, щоб одна з матриць була транспонована.
# Так як жодна з матриць не є транспонованною ми отримуємо помилку.
res = np.dot(a, b)
print(res)
````

````python
# Пояснення: Добуток Адамара є операцією, що здійснюється над двома матрицями або векторами однакових розмірностей.
# Він обчислюється шляхом множення відповідних елементів обох матриць або векторів.
res = a * b
print(res)
````

````python
# Пояснення: Обчислюємо шляхом ділення відповідних елементів обох матриць або векторів.
res = a / b
print(res)
````

````python
# Пояснення: Кожен елемент рядкового вектора а ділимо на стовпчик b_T
res = a / b.T
print(res)
````

#### 2.1 Зменши вектор  $x$  в 2 рази по вісі  OX  та збільш в 3 рази по вісі  OY .

$$
M_1 =
\left (\begin{array}{cc}

$$

$$
